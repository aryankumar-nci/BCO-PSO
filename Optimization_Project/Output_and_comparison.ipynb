```python
def Fun1(solution,size1):
  list1 = [0] * data['vmcount'][0]
  tid=[]
  vid=[]
  start_t=[]
  finish_t=[]
  exe_t=[]
  val = 0.0
  #print(solution)
  #print(int(round(solution[0])))
  for i in range(len(solution)):
    solution[i]=round(solution[i])
    if(round(solution[i])!=0.0):
      start=list1[int(abs(round(solution[i],0)))-1]
      list1[int(abs(round(solution[i],0)))-1]=list1[int(abs(round(solution[i],0)))-1] + data['task'][0]/data['VM'][int(abs(round(solution[i],0)))-1]
      finish=list1[int(abs(round(solution[i],0)))-1]
      tid.append(i)
      vid.append(abs(round(solution[i],0)))
      start_t.append(start)
      finish_t.append(finish)
      exe_t.append(finish-start)

    else:
      list1[int(abs(round(solution[i])))]=list1[int(abs(round(solution[i],0)))] + data['task'][0]/data['VM'][int(abs(round(solution[i])))]
      tid.append(i)
      vid.append(abs(round(solution[i],0)))
      start_t.append(start)
      finish_t.append(finish)
      exe_t.append(finish-start)
  print(max(list1))
  l=max(list1)
  uitli=[x/l for x in list1]
  import numpy as np
  utlim=np.mean(uitli)*100
  print(len(tid))
  simula={'task_id': tid, 'VM_ID': vid, 'Start_time': start_t,'Finish_time':finish_t,'Exec_time':exe_t}
  df=pd.DataFrame(simula)
  print(df)

  return l, utlim,df
```
